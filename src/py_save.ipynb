hfaデータは時間帯によってチャンネルが違うので、lowとmonitデータで補完

```python
timelow = pytplot.data_quants['erg_pwe_hfa_l2_low_spectra_e_mix']['time']
low = pytplot.data_quants['erg_pwe_hfa_l2_low_spectra_e_mix']
lowb = pytplot.data_quants['erg_pwe_hfa_l2_low_spectra_bgamma']
timemonit = pytplot.data_quants['erg_pwe_hfa_l2_monit_spectra_e_mix']['time']
monit = pytplot.data_quants['erg_pwe_hfa_l2_monit_spectra_e_mix']
monitb = pytplot.data_quants['erg_pwe_hfa_l2_monit_spectra_bgamma']
timelow = timelow.to_numpy()
timemonit = timemonit.to_numpy()
len_v_dim = len(pytplot.data_quants['erg_pwe_hfa_l2_monit_spectra_e_mix']['v_dim'])
len_v_dim_b = len(pytplot.data_quants['erg_pwe_hfa_l2_monit_spectra_bgamma']['v_dim'])
v_dim = pytplot.data_quants['erg_pwe_hfa_l2_monit_spectra_e_mix']['v_dim']
v_dim_b = pytplot.data_quants['erg_pwe_hfa_l2_monit_spectra_bgamma']['v_dim']
timemerged = np.sort(np.hstack((timelow, timemonit)))
merged_data = np.zeros((len(timemerged), len_v_dim))
merged_datab = np.zeros((len(timemerged), len_v_dim_b))

# 'low' データを 'timelow' に対応する位置にコピー
for i, timestamp in enumerate(timelow):
    index = np.where(timemerged == timestamp)[0]
    merged_data[index] = low[i]

# 'omit' データを 'timeomit' に対応する位置にコピー
for i, timestamp in enumerate(timemonit):
    #mask = timemerged == timestamp  # 'timestamp' と一致する場所を示すブール値の配列を作成
    index = np.where(timemerged == timestamp)[0]
    merged_data[index] = monit[i]

for i, timestamp in enumerate(timelow):
    index = np.where(timemerged == timestamp)[0]
    merged_datab[index] = lowb[i]

for i, timestamp in enumerate(timemonit):
    index = np.where(timemerged == timestamp)[0]
    merged_datab[index] = monitb[i]

pytplot.store_data('merged', data={'x': timemerged, 'y': merged_data, 'v': v_dim})
pytplot.store_data('mergedb', data={'x': timemerged, 'y': merged_datab, 'v': v_dim_b})

```

l2のポテンシャルデータはdsiでxy方向になっているので、uとvを見るためにノルムをとる

```python
Eux = pytplot.data_quants['erg_pwe_efd_l2_E_spin_Eu_dsi'][:,0]
Euy = pytplot.data_quants['erg_pwe_efd_l2_E_spin_Eu_dsi'][:,1]
Evx = pytplot.data_quants['erg_pwe_efd_l2_E_spin_Ev_dsi'][:,0]
Evy = pytplot.data_quants['erg_pwe_efd_l2_E_spin_Ev_dsi'][:,1]
Euoff = pytplot.data_quants['erg_pwe_efd_l2_E_spin_Eu_offset']
Evoff = pytplot.data_quants['erg_pwe_efd_l2_E_spin_Ev_offset']
EuEvoff = np.zeros((len(Euoff), 2))
EuEvoff[:,0] = Euoff
EuEvoff[:,1] = Evoff
Eu = np.sqrt(Eux**2 + Euy**2)
Ev = np.sqrt(Evx**2 + Evy**2)
EuEv = np.zeros((len(Eu), 2))
EuEv[:,0] = Eu
EuEv[:,1] = Ev
pytplot.store_data('EuEv', data={'x': pytplot.data_quants['erg_pwe_efd_l2_E_spin_Eu_dsi']['time'], 'y': EuEv})
pytplot.store_data('EuEvoff', data={'x': pytplot.data_quants['erg_pwe_efd_l2_E_spin_Eu_dsi']['time'], 'y': EuEvoff})
```

プロットオプションを設定

```python
labels = pytplot.split_vec( 'erg_orb_l2_pos_rmlatmlt' )
pytplot.options( 'erg_orb_l2_pos_rmlatmlt_x', opt_dict={'ytitle':'L [Re]','ytitle_location':'left'} )
pytplot.options( 'erg_orb_l2_pos_rmlatmlt_y', 'ytitle', 'MLat [deg]' )
pytplot.options( 'erg_orb_l2_pos_rmlatmlt_z', 'ytitle', 'MLT [h]' )
pytplot.options('EuEv', opt_dict={'legend_names': ['u', 'v'], 'yrange': [ 0, 10], 'ytitle':'EFD amplitude', 'ysubtitle':'[mV/m]'})
pytplot.options('EuEvoff', opt_dict={'legend_names': ['u', 'v'], 'yrange': [ -10,10], 'ytitle':'EFD offset', 'ysubtitle':'[mV/m]', 'ztitle':'[$mV^2/m^2/Hz$]'})
pytplot.options('merged', opt_dict={'zlog': True, 'ylog': True, 'yrange': [1e-0, 1e4], 'ysubtitle':'[kHz]', 'zrange': [1e-10, 1e-3], 'spec': True, 'ytitle':'HFA Emix', 'ztitle':'[$mV^2/m^2/Hz$]'})
pytplot.options( 'mergedb', opt_dict={ 'ytitle':'HFA(Bz)', 'ysubtitle':'[kHz]', 'yrange':[2, 200], 'ztitle':'[$pT^2/Hz$]', 'spec': True, 'zlog': True, 'ylog': True})
pytplot.options( 'erg_pwe_ofa_l2_spec_E_spectra_132' , opt_dict={'zlog': True, 'ylog': True, 'zrange': [1e-9, 1e-2], 'spec': True, 'ytitle':'OFA-E', 'ysubtitle':'[kHz]', 'ztitle':'[$mV^2/m^2/Hz$]'})
pytplot.options( 'erg_pwe_ofa_l2_spec_B_spectra_132' , opt_dict={'zlog': True, 'ylog': True, 'zrange': [1e-4, 1e2], 'spec': True, 'ytitle':'OFA-B', 'ysubtitle':'[kHz]', 'ztitle':'[$pT^2/Hz$]'})
pytplot.options('erg_pwe_efd_l2_spec_spectra', opt_dict={'zlog': True, 'ylog': 0, 'zrange': [1e-7, 1e-0], 'spec': True, 'ytitle':'EFD', 'ysubtitle':'[Hz]'})
pytplot.options('erg_pwe_efd_l2_pot_Vave', opt_dict={'yrange':[-10,10], 'ytitle':'potential', 'ysubtitle':'[V]'})
```
